```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
```

```python
wavelengths = np.linspace(2E-7, 2E-5, 500, dtype=np.float128)
def planck(L, T):
    h = 6.6261E-34
    kB = 1.381E-24
    c = 3E8

    return 2*h*c**2/L**5 * 1/(np.exp(h*c/(L*kB*T)) - 1)

data = planck(wavelengths, 8000) + np.random.normal(size=wavelengths.size)*.1E9
```

```python
plt.plot(wavelengths, data, '.')
```

```python
pfit, pcov = curve_fit(planck, xdata=wavelengths, ydata=data, p0=[1000])
```

```python
plt.plot(wavelengths, data, '.')
plt.plot(wavelengths, planck(wavelengths, pfit[0]))
```
