```python
import matplotlib.pyplot as plt
```

```python
ras = []
decs = []
with open('./brightest_200.csv') as fh:
    for line in fh:
        entry = line.strip().split(',')
        ra = entry[-2]
        dec = entry[-1]
        ra_deg = (float(ra[0:2]) + float(ra[3:5])/60) * 15
        dec_deg = float(dec)
        ras.append(ra_deg)
        decs.append(dec_deg)

plt.subplot(111, projection='aitoff')
plt.plot(ras, decs, '.')
plt.xlabel('Right Ascension (deg)')
plt.ylabel('Declination (deg)')
plt.grid(True)
plt.show()
```
